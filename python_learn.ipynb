```python
# Write code  💖
Gryffindor = 0
Ravenclaw = 0
Hufflepuff = 0
Slytherin = 0

# --- Question 1 ---
print("Q1) Do you like Dawn or Dusk?")
print("    1) Dawn")
print("    2) Dusk")
Q1 = int(input("Enter answer (1-2): "))

if Q1 == 1:
    Gryffindor += 1
    Ravenclaw += 1
elif Q1 == 2:
    Hufflepuff += 1
    Slytherin += 1
else:
    print("Wrong input.")

# --- Question 2 ---
print("\nQ2) When I’m dead, I want people to remember me as:")
print("    1) The Good")
print("    2) The Great")
print("    3) The Wise")
print("    4) The Bold")
Q2 = int(input("Enter answer (1-4): "))

if Q2 == 1:
    Hufflepuff += 2
elif Q2 == 2:
    Slytherin += 2
elif Q2 == 3:
    Ravenclaw += 2
elif Q2 == 4:
    Gryffindor += 2
else:
    print("Wrong input.")

# --- Question 3 ---
print("\nQ3) Which kind of instrument most pleases your ear?")
print("    1) The violin")
print("    2) The trumpet")
print("    3) The piano")
print("    4) The drum")
Q3 = int(input("Enter answer (1-4): "))

if Q3 == 1:
    Slytherin += 4
elif Q3 == 2:
    Hufflepuff += 4
elif Q3 == 3:
    Ravenclaw += 4
elif Q3 == 4:
    Gryffindor += 4
else:
    print("Wrong input.")

# --- Final Scores ---
print(f"""\nAmount of points each House:
        Ravenclaw = {Ravenclaw}
        Gryffindor = {Gryffindor}
        Hufflepuff = {Hufflepuff}
        Slytherin = {Slytherin} """)

# --- Bonus: Find the house with the most points ---
# Menyimpan skor dalam dictionary agar mudah dicari nilai tertingginya
houses = {
    "Gryffindor": Gryffindor,
    "Ravenclaw": Ravenclaw,
    "Hufflepuff": Hufflepuff,
    "Slytherin": Slytherin
}

# Fungsi max() akan mencari nilai tertinggi, lalu mengambil kunci (nama asrama) dari nilai tersebut
winner = max(houses, key=houses.get)
loser = min(houses, key=houses.get)

print(f"\n🎉 Congratulations! The Sorting Hat has chosen: {winner}! 🎉")
print(f"\nkamu tidak cocok di houses {loser}")
```
